```python
#AUTORES: JULIÁN EDUARDO REYES CALDERÓN - JUAN SEBASTIAN LARROTA

import pandas as pd
import numpy as np
import tensorflow as tf
import csv
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Definir la ruta del archivo
ruta_archivo = "/content/Dataset/Diagnostico(Datos).csv"

# Verificar contenido del archivo antes de cargarlo
with open(ruta_archivo, "r", encoding="utf-8") as f:
    primeras_lineas = [next(f) for _ in range(5)]
    print("Primeras líneas del archivo CSV:")
    for linea in primeras_lineas:
        print(linea.strip())

# Detectar automáticamente el delimitador
with open(ruta_archivo, "r", encoding="utf-8") as f:
    dialecto = csv.Sniffer().sniff(f.read(1024))
    delimitador = dialecto.delimiter
    print(f"Delimitador detectado: '{delimitador}'")

# Cargar el archivo CSV con el delimitador detectado
df = pd.read_csv(ruta_archivo, sep=delimitador, encoding="utf-8", engine="python", on_bad_lines="skip")

# Reemplazar comas por puntos en valores numéricos
df = df.apply(lambda x: x.map(lambda y: str(y).replace(",", ".") if isinstance(y, str) else y))

# Convertir valores de texto en numéricos (Ejemplo: 'GENERO')
if 'GENERO' in df.columns:
    df['GENERO'] = df['GENERO'].map({'MASCULINO': 0, 'FEMENINO': 1})

# Convertir todas las columnas posibles a tipo numérico
df = df.apply(pd.to_numeric, errors="coerce")

# Imputar valores nulos con la mediana solo en columnas numéricas
columnas_numericas = df.select_dtypes(include=[np.number]).columns
df[columnas_numericas] = df[columnas_numericas].fillna(df[columnas_numericas].median())

# Verificar los nombres de las columnas
print("Columnas en el DataFrame:", df.columns)

# Seleccionar características de entrada y etiqueta
if 'EVOLUCION' in df.columns:
    X = df.drop(columns=['EVOLUCION'])
    y = df['EVOLUCION']

    # Dividir el dataset (80/20) sin aleatorización
    train_size = int(0.8 * len(df))
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    # Definir el modelo
    model = Sequential([
        Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
        Dense(32, activation='relu'),
        Dense(1)  # Capa de salida para regresión
    ])

    # Compilar el modelo
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='mean_squared_error')

    # Entrenar el modelo
    history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))

    # Evaluar el modelo
    loss = model.evaluate(X_test, y_test)
    print(f'Pérdida en el conjunto de prueba: {loss}')

    # Mostrar estadísticas después de la transformación
    print("\nEstadísticas después de la normalización/estandarización:")
    print(df.describe())
else:
    print("La columna 'EVOLUCION' no se encuentra en el DataFrame.")
```
